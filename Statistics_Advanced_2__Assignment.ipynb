1. What is hypothesis testing in statistics?
 -  Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves making an initial assumption called the null hypothesis (H‚ÇÄ), which usually represents no effect, no difference, or a status quo. An alternative hypothesis (H‚ÇÅ or Ha) is then proposed, suggesting there is a significant effect or difference. The goal is to determine whether the sample data provides enough evidence to reject the null hypothesis in favor of the alternative. This is done by calculating a test statistic and comparing it to a threshold determined by the significance level (commonly 0.05). A p-value is also computed, which indicates the probability of obtaining the observed results if the null hypothesis were true. If the p-value is less than the significance level, the null hypothesis is rejected. Hypothesis testing helps researchers make objective decisions and is widely used in fields like medicine, business, and social sciences to evaluate claims or test theories.

2.  What is the null hypothesis, and how does it differ from the alternative
hypothesis?
  -  The null hypothesis, denoted as H‚ÇÄ, is a foundational concept in hypothesis testing and represents the default or initial assumption that there is no effect, no difference, or no relationship between variables. It is the hypothesis that researchers seek to test and possibly disprove. In contrast, the alternative hypothesis, denoted as H‚ÇÅ or Ha, represents what the researcher aims to support‚Äîit suggests that there is a significant effect, difference, or relationship. The null hypothesis is assumed true unless there is strong statistical evidence to reject it, while the alternative hypothesis is considered only if the data provide enough support against the null. In essence, the null hypothesis is a statement of "no change" or "no difference," whereas the alternative hypothesis reflects a new theory or claim that contradicts the null.

3. Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
 - The significance level, often denoted by the Greek letter alpha (Œ±), is a threshold set by the researcher before conducting a hypothesis test. It represents the probability of making a **Type I error**, which occurs when the null hypothesis (H‚ÇÄ) is wrongly rejected even though it is actually true. Common significance levels are 0.05, 0.01, or 0.10, with 0.05 being the most widely used. This means there is a 5% risk of rejecting the null hypothesis when it is correct. The significance level plays a crucial role in deciding the outcome of a test by serving as a benchmark for the **p-value**, which is the probability of observing the sample data (or more extreme) under the assumption that the null hypothesis is true. If the p-value is **less than or equal to Œ±**, the null hypothesis is rejected, indicating that the observed result is statistically significant. If the p-value is **greater than Œ±**, the null hypothesis is not rejected, suggesting that there is not enough evidence to support the alternative hypothesis. Therefore, the significance level helps determine whether the test results are strong enough to conclude a real effect or difference.

4. What are Type I and Type II errors? Give examples of each.
 - In hypothesis testing, **Type I and Type II errors** are two kinds of incorrect conclusions that can result from making decisions based on sample data.

A **Type I error** occurs when the **null hypothesis (H‚ÇÄ) is true**, but we **incorrectly reject it**. In other words, we conclude that there is an effect or difference when there actually isn't one. The probability of making a Type I error is equal to the **significance level (Œ±)**.
**Example**: Suppose a medical test is used to detect a disease. If a healthy person is wrongly diagnosed as having the disease, that‚Äôs a **Type I error**‚Äîrejecting the null hypothesis that the person is healthy when they actually are.

A **Type II error** occurs when the **null hypothesis is false**, but we **fail to reject it**. This means we miss detecting an actual effect or difference. The probability of making a Type II error is denoted by **Œ≤**.
**Example**: Using the same medical test, if a person who actually has the disease is told they are healthy, that‚Äôs a **Type II error**‚Äîfailing to reject the null hypothesis of being disease-free when it is false.

In summary, a **Type I error** is a "false positive," and a **Type II error** is a "false negative." Both errors have serious implications depending on the context, so choosing appropriate significance levels and test designs is important to minimize them.

5. What is the difference between a Z-test and a T-test? Explain when to use
each
 - The main difference between a **Z-test** and a **T-test** lies in the type of data you have and the **sample size**. Both are used to determine whether there is a significant difference between means, but they are applied under different conditions.

A **Z-test** is used when the **population standard deviation is known** and the **sample size is large** (typically **n ‚â• 30**). It relies on the assumption that the sampling distribution of the mean is normal, which is more likely with larger samples due to the Central Limit Theorem. For example, if you're comparing the sample mean to a known population mean and you know the population standard deviation, a Z-test is appropriate.

On the other hand, a **T-test** is used when the **population standard deviation is unknown** and the **sample size is small** (typically **n < 30**). Instead of using the normal distribution, it uses the **t-distribution**, which has heavier tails to account for the extra uncertainty in estimating the population standard deviation from the sample. A T-test is commonly used in real-world scenarios, such as comparing the average test scores of two small classes or measuring the effect of a treatment on a small group of patients.

In summary, use a **Z-test** when you have a large sample and know the population standard deviation; use a **T-test** when you have a small sample and do not know the population standard deviation.

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu_0 = 50               # hypothesized population mean
sigma = 0.5             # assumed population standard deviation
alpha = 0.05            # significance level

# Sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Z-statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed test: calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Output results
print(f"Sample Mean = {sample_mean:.4f}")
print(f"Z-statistic = {z:.4f}")
print(f"P-value = {p_value:.4f}")

# Decision
if p_value < alpha:
    print("Reject the null hypothesis (H‚ÇÄ).")
else:
    print("Fail to reject the null hypothesis (H‚ÇÄ).")
```

8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
#ans 8
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility
mean = 100
std_dev = 15
sample_size = 100

data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

# Step 2: Calculate 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
confidence_level = 0.95
alpha = 1 - confidence_level

# Calculate margin of error using t-distribution
t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 3: Plot the data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
plt.axvline(ci_lower, color='green', linestyle='--', label=f'Lower 95% CI = {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='--', label=f'Upper 95% CI = {ci_upper:.2f}')
plt.title("Histogram of Simulated Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

Z-scores allow comparison between different datasets or distributions

```python
#ans 9 import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Step 1: Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data, ddof=0)  # Population standard deviation

    # Step 2: Calculate Z-scores
    z_scores = [(x - mean) / std_dev for x in data]

    # Step 3: Plot histogram of Z-scores
    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score (Standard Deviations from Mean)")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.legend()
    plt.show()

    # Step 4: Return Z-scores
    return z_scores

# Example usage
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

z_scores = calculate_and_plot_z_scores(data)

# Print first 5 Z-scores as a sample
print("First 5 Z-scores:", z_scores[:5])
```

Interpretation of Z-scores:

A Z-score indicates how far (and in which direction) a value is from the mean, in units of standard deviation.

For example:

Z = +2.0 ‚Üí the value is 2 std devs above the mean.

Z = -1.5 ‚Üí the value is 1.5 std devs below the mean.

This standardization allows comparisons across datasets with different scales or units.
